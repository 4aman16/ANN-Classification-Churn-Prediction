```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np

```

```python
## Load the Ann train model and one hot encodng and scaler pickle file

model= load_model('model.h5')

with open('label_encoder.pkl','rb') as file:
    label_encoder= pickle.load(file)

with open('onehot_encoder_geo.pkl','rb') as file:
    one_hot= pickle.load(file)

with open('scaler.pkl','rb') as file:
    scaler= pickle.load(file)
```

```python
input_data={
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
#Convert the input data to numerical values
# use one hot encoder
one_hot_encode_geo = one_hot.transform([[input_data['Geography']]])
one_hot_encode_geo_df=pd.DataFrame(one_hot_encode_geo.toarray(), columns=one_hot.get_feature_names_out(['Geography']))
```

```python
one_hot_encode_geo_df
input_df=pd.DataFrame([input_data])
input_df
```

```python
## Convert Categorical value as well

input_df['Gender']=label_encoder.transform(input_df['Gender'])
```

```python
input_df
```

```python
input_df= pd.concat([input_df.drop('Geography',axis=1),one_hot_encode_geo_df],axis=1)
```

```python
input_df
```

```python
## SCaling teh input data

input_scaled= scaler.transform(input_df)
```

```python
input_scaled
```

```python
## Prdiction

prediction = model.predict(input_scaled)
prediction
```
